### Setup App Structure for Receipt Generator

```python
# Import external tools needed
from PIL import Image, ImageDraw, ImageFont

# storage
cart = {}

def add_item(name,price,quantity):
      cart[name] = {'price': price, 'quantity': quantity}

# loop
while True:
    item = input('Enter Product Name:')
    if item.lower() == 'done':
        break
    price = float(input('Enter Price:'))
    quantity = int(input('Enter Quantity:'))
    add_item(item, price, quantity)

# calculate Total
def calculate_total():
    total = 0
    for name in cart:
        price = cart[name]['price'] 
        quantity = cart[name]['quantity']
        total += price * quantity
    return total

# Apply Discount
# for any sales greater that 500, give a 10% discount
def apply_discount(total):
    if total > 500:
        return total * 0.9 
    return total

# tax is 7.5%
def taxifier(apply_discount):
    tax_amt = apply_discount * 0.075
    return tax_amt

# Tax calculation
def tax_calc(total):
     return total * 1.075

# Save Receipt
def save_receipt(lines, filename='reciept.png'):
    
    # input filename
    filename=f'{input('Enter Transaction Description:')}.png'
    
    # file size
    width = 400
    line_height = 25
    height = line_height * (len(lines) + 2)
    
    #create imaage
    image = Image.new('RGB', (width,height), 'white')
    draw= ImageDraw.Draw(image)
    font = ImageFont.truetype('arial.ttf', 16)

    # Draw each line 
    y = 10
    for line in lines:
        draw.text((10, y), line, fill="black", font=font)
        y += line_height
    image.save(filename)
    print(f'Receipt saved as {filename}')

# Print_receipt
def print_receipt():
    # storage for lines list    
    lines = ['------Receipt------']

    for item, detail in cart.items():
        lines.append(f"{item.upper()}: {detail['quantity']} x {detail['price']} = {detail['quantity'] * detail['price']}")
        
    lines.append(f"DISCOUNT of 10%: ${apply_discount(calculate_total()):.2f}")
    lines.append(f"TAX of 7.5%: ${taxifier(apply_discount(calculate_total())):.2f}")
    lines.append(f"TOTAL: ${tax_calc(apply_discount(calculate_total())):.2f}")

    print(f"DISCOUNT of 10%: ${apply_discount(calculate_total()):.2f}")
    print(f"TAX of 7.5%: ${taxifier(apply_discount(calculate_total())):.2f}")
    print(f"TOTAL: ${tax_calc(apply_discount(calculate_total())):.2f}")

    save_receipt(lines)
    
print_receipt()
```
